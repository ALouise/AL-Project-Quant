# SCIPY


```python
#fonction d'interpolation
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d 

x = np.linspace(0,10,10)
y =x**3 + 3*x
plt.scatter(x,y)
f = interp1d(x,y, kind='cubic')
new_x = np.linspace(0,10,30)
result = f(new_x)
plt.plot(new_x, result, c='r') 
```
